```python
import matplotlib.pyplot as plt
import seaborn as sns

# Asegurarse que '1º actividad' es datetime
datos["1º actividad"] = pd.to_datetime(datos["1º actividad"], errors="coerce")

# Crear columna de mes-año
datos["Mes"] = datos["1º actividad"].dt.to_period("M")

# Contar FO por Coordinador y mes
fo_por_mes = datos.groupby(["Coordinador", "Mes"])["FO"].count().reset_index(name="CantidadFO")

# Calcular promedio de FO por mes para cada Coordinador
promedio_fo = fo_por_mes.groupby("Coordinador")["CantidadFO"].mean().reset_index()

# Ordenar de mayor a menor promedio
promedio_fo = promedio_fo.sort_values(by="CantidadFO", ascending=False)

# Estilo de Seaborn
sns.set_theme(style="whitegrid")

# Crear gráfico de barras
plt.figure(figsize=(12,7))
barplot = sns.barplot(
    x="Coordinador",
    y="CantidadFO",
    data=promedio_fo,
    palette="Blues_d"
)

# Etiquetas y título
plt.xlabel("Coordinador", fontsize=12)
plt.ylabel("Promedio de FO por mes", fontsize=12)
plt.title("Promedio mensual de FO por Coordinador", fontsize=16, weight='bold')

# Mostrar valores encima de cada barra
for index, row in enumerate(promedio_fo.itertuples()):
    barplot.text(index, row.CantidadFO + 0.2, f"{row.CantidadFO:.1f}", ha='center', fontsize=10)

# Mejorar visibilidad del eje x
plt.xticks(rotation=45, ha='right', fontsize=11)
plt.yticks(fontsize=11)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Asegurarse que '1º actividad' es datetime
datos["1º actividad"] = pd.to_datetime(datos["1º actividad"], errors="coerce")

# Filtrar solo meses mayo, junio, julio
datos_filtrados = datos[datos["1º actividad"].dt.month.isin([5,6,7])]

# Crear columna de mes con nombre corto
datos_filtrados["Mes"] = datos_filtrados["1º actividad"].dt.strftime("%b")  # 'May', 'Jun', 'Jul'

# Contar FO por Coordinador y Mes
fo_por_mes = datos_filtrados.groupby(["Coordinador", "Mes"])["FO"].count().reset_index(name="CantidadFO")

# Ordenar los meses para que queden May, Jun, Jul
mes_orden = ["May", "Jun", "Jul"]
fo_por_mes["Mes"] = pd.Categorical(fo_por_mes["Mes"], categories=mes_orden, ordered=True)

# Gráfico de barras agrupadas
plt.figure(figsize=(14,7))
sns.set_theme(style="whitegrid")
sns.barplot(
    x="Coordinador",
    y="CantidadFO",
    hue="Mes",
    data=fo_por_mes,
    palette="Set2"
)

plt.xlabel("Coordinador", fontsize=12)
plt.ylabel("Cantidad de FO", fontsize=12)
plt.title("FO por Coordinador (Mayo, Junio, Julio)", fontsize=16, weight='bold')
plt.legend(title="Mes")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Asegurarse que '1º actividad' es datetime
datos["1º actividad"] = pd.to_datetime(datos["1º actividad"], errors="coerce")

# Filtrar solo meses mayo, junio, julio
datos_filtrados = datos[datos["1º actividad"].dt.month.isin([5,6,7])]

# Crear columna de mes con nombre corto
datos_filtrados["Mes"] = datos_filtrados["1º actividad"].dt.strftime("%b")  # 'May', 'Jun', 'Jul'

# Contar FO por Coordinador y Mes
fo_por_mes = datos_filtrados.groupby(["Coordinador", "Mes"])["FO"].count().reset_index(name="CantidadFO")

# Ordenar los meses para que queden May, Jun, Jul
mes_orden = ["May", "Jun", "Jul"]
fo_por_mes["Mes"] = pd.Categorical(fo_por_mes["Mes"], categories=mes_orden, ordered=True)

# Gráfico de barras agrupadas
plt.figure(figsize=(14,7))
sns.set_theme(style="whitegrid")
ax = sns.barplot(
    x="Coordinador",
    y="CantidadFO",
    hue="Mes",
    data=fo_por_mes,
    palette="Set2"
)

# Agregar los valores encima de cada barra
for p in ax.patches:
    height = p.get_height()
    ax.annotate(f'{int(height)}',
                (p.get_x() + p.get_width() / 2., height),
                ha='center', va='bottom', fontsize=10, fontweight='bold')

plt.xlabel("Coordinador", fontsize=12)
plt.ylabel("Número de FO", fontsize=12)
plt.title("Número de FO por Coordinador y Mes (Mayo, Junio, Julio)", fontsize=16, weight='bold')
plt.legend(title="Mes")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
